```python
import pandas as pd

# Step 1: Change "Date" to Pandas datetime format
df_t['Date'] = pd.to_datetime(df_t['Date'])
df_s['Date'] = pd.to_datetime(df_s['Date'])

# Step 2: Sort both dataframes by "Date"
df_t = df_t.sort_values(by='Date')
df_s = df_s.sort_values(by='Date')
```

```python
df_t.head()
```

```python
df_s.drop(columns=["Date Nepali"],inplace=True)
```

```python
df_s.head()
```

```python

# Step 3: Merge the two dataframes based on the "Date" column (left join)
merged_df = pd.merge(df_t, df_s, on='Date', how='left')
```

```python
merged_df.isnull().sum()
```

```python
merged_df.head()
```

```python
x=merged_df.columns
```

```python
# Find the first non-NaN value in each column
first_non_nan = merged_df.ffill().iloc[-1]

# Replace NaN values in columns 0 to i-1 with the first non-NaN value in the corresponding column
for col in range(merged_df.shape[1]):
    merged_df.iloc[:, col] = merged_df.iloc[:, col].fillna(first_non_nan[col])
```

```python
print(merged_df.head())
```

```python
columns_to_fill = ["EPS","Book Value", "COF","Dividend"]  # Add columns to fill here
for column in columns_to_fill:
    merged_df[column].fillna(method='ffill', inplace=True)
```
