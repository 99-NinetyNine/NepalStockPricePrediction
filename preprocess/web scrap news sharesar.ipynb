```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import time


# Initialize the Firefox WebDriver
driver = webdriver.Chrome()

# Open a website

url = 'https://www.sharesansar.com/company/nabil'  # Replace with your desired URL
driver.get(url)

# Initialize an empty list to store table data
all_table_data = []
i=80
# Find and click the link with text "News"
driver.find_element(By.ID, 'btn_cnews').click()
time.sleep(1)

while i>0:
    i-=1
    
    # Wait for 1 second
    
    # Find the table with ID "myTableCNews"
    table = driver.find_element(By.ID, 'myTableCNews')
    
    # Extract the table content as HTML
    table_content = table.get_attribute('outerHTML')
    
    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(table_content, 'html.parser')
    
    # Extract and process the table rows (adjust this part according to your table structure)
    rows = soup.find_all('tr')
    
    for row in rows:
        columns = row.find_all('td')
        row_data =[]
        
        for column in columns:
            row_data.append(column.get_text())
            a_tag = column.find('a')
            if a_tag:
                row_data.append(a_tag.get('href'))  
            
        
        all_table_data.append(row_data)
    # Find and click the "Next" button with ID "myTableCNews_next"
    next_button = driver.find_element(By.ID, 'myTableCNews_next')
    next_button.click()
    time.sleep(3)
    print("hi")
    # Wait for 1 second (adjust the sleep time as needed)
    
    
    
# Close the browser
driver.quit()

# Now, all_table_data contains the table data as a list of lists
# You can append or process this data as needed
for row in all_table_data:
    print(row)  # Print each row of data
```

```python
detail_news=[]
i=1

from selenium.webdriver.chrome.options import Options
# Initialize the WebDriver
chrome_options = Options()
chrome_options.add_argument("--headless")
driver = webdriver.Chrome(options=chrome_options)

for row in all_table_data:
    whole_paragraph = ""
    try:
        url=row[2]
        
        # Initialize an empty string to store the combined paragraphs
        driver.get(url)
        div_elements = driver.find_elements(By.CLASS_NAME, "col-md-12")
        
        
        # Iterate through the div elements
        for div_element in div_elements:
            # Find all <p> elements with style "text-align: justify;" within the div
            p_elements = div_element.find_elements(By.CSS_SELECTOR, 'p[style="text-align: justify;"]')
            
            # Extract text from each <p> element and add it to the whole_paragraph
            for p in p_elements:
                whole_paragraph += p.text + " "
        time.sleep(2)
        print("say hello")
    except:
        print("say hi")
        
    print("done",i)
    i+=1
    detail_news.append(whole_paragraph)
    
driver.quit()
```

```python
dn=pd.DataFrame(detail_news)
c_df=pd.concat([df,dn], axis=1)


# Save the DataFrame as a CSV file
c_df.to_csv('my_dataframe.csv', index=False)
```
